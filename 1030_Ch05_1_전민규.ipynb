```python
img = cv2.imread('/content/gdrive/My Drive/IP/Images/lena256.jpg')
print(img.shape)
gImg = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
cv2_imshow(gImg)
```

```python
blurImage = cv2.GaussianBlur(gImg, (7,7), 1)
cv2_imshow(blurImage)
```

```python
edgeImage = cv2.addWeighted(gImg, 1, blurImage, -1, 0)
cv2_imshow(edgeImage)
```

```python
unsharpImage = cv2.addWeighted(gImg, 1, edgeImage, 1, 0)
cv2_imshow(unsharpImage)
```
